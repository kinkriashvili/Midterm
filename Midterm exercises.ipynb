# სავარჯიშო 4 - სტატისტიკა (5 ქულა)
### კუმულაციური ალბათობა (Cumulative distribution function)
დაითვალეთ შემდეგი კუმულაციური ალბათობა scipy.stats.norm.cdf() ფუნქციის გამოყენებით. დაითვალეთ მილიარდელების ჩამონათვლიდან რამდენ პროცენტია 50 წლის ზემოთ. 
norm.cdf(x, mean_val, std_dev_val) ფუნქციის პირველი პარამეტრია სასაზღვრო მნიშვნელობა, მეორე - საშუალო მნიშვნელობა, მესამე -სტდ. გადახრა.

მოახდინეთ შედეგების სიტყვიერი ინტერპრეტაცია.

იხილეთ დოკუმენტაციის ლინკი: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html

```python
from scipy.stats import norm
```

```python
# border = 50 - განვსაზღვრე ასაკი - 50, რომელიც გამოყენებული იქნება ამ ასაკის ზემოთ მილიარდელების დასათვლელად
# mean = df['age'].mean() - ითვლის მილიარდერების საშუალო ასაკს
# std_dev = df['age'].std() - ითვლის მონაცემთა ნაკრებში მილიარდერების ასაკის სტანდარტულ გადახრას.


border = 50
mean = df['age'].mean()
std_dev = df['age'].std()

# cumulative_prob = 1 - norm.cdf(border, mean, std_dev) - ითვლის კუმულატიურ ალბათობას norm.cdf() ფუნქციის გამოყენებით scipy.stats მოდულიდან.

# სასაზღვრო მნიშვნელობის (50 წლის ასაკი), საშუალო და სტანდარტული გადახრის გადაცემით (norm.cdf(border, mean, std_dev)) 
# ვღებულობთ ალბათობას, რომ ამ განაწილებიდან შემთხვევითი ნიმუში ნაკლებია ან ტოლია საზღვრის მნიშვნელობაზე. 

# ამ ალბათობის გამოკლება 1-დან მოგცემთ კუმულატიურ ალბათობას, რომ იყოთ საზღვრის მნიშვნელობაზე (ამ შემთხვევაში, 50 წელზე მეტი ასაკის).
cumulative_prob = 1 - norm.cdf(border, mean, std_dev)

# გადავაქცევ კუმულატიურ ალბათობას პროცენტში მისი 100-ზე გამრავლებით.
percentage = cumulative_prob * 100

print(f"Percentage of billionaires over the age of 50: {percentage:.2f}%")
```

# სავარჯიშო 6 - Numpy (5 ქულა)

 შეასრულეთ შემდეგი დავალებებლი numpy ბიბლიოთეკის გამოყენებით:
 1. შექმენით 3x3-ზე A მატრიცა, რომელიც შედგენილი იქნება 0-დან 10-მდე შემთხვევითი მთელი რიცხვებისგან.
 2. ასევე შექმენით ერთეულოვანი (1-ებისგან შედგენილი) B მატრიცა.
 3. გამოთვალეთ AxB (ამ ორი მატრიცის ნამრავლი) matmul ფუნქციის გამოყენებით.
 4. იპოვეთ A მატრიცის მინიმალური ელემენტის მნიშვნელობა. 
 
 იხილეთ დოკუმენტაცია https://numpy.org/doc/stable/reference/generated/numpy.matmul.html

```python
import numpy as np
```

```python
# კოდი ქმნის 3x3 მატრიცას A, რომელიც სავსეა შემთხვევითი მთელი რიცხვებით 0-დან 10-მდე np.random.randint() გამოყენებით.
# ქმნის 3x3 უნიტარულ მატრიცას B, რომელიც ივსება 1-ებით np.ones() გამოყენებით.
# ითვლის A და B მატრიცების ნამრავლს np.matmul() ფუნქციის გამოყენებით.
# პოულობს მინიმალურ მნიშვნელობას A მატრიცაში np.min() ფუნქციის გამოყენებით.

A = np.random.randint(0, 11, size=(3, 3))


B = np.ones((3, 3))

result = np.matmul(A, B)

min_value_A = np.min(A)

print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
print("\nResult of AxB:")
print(result)
print("\nMinimum value of matrix A:", min_value_A)
```
